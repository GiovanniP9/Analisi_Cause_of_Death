```python

# Filtra per un Paese (es. "Italy")
country = "Italy"
df_country = df[df["Country/Territory"] == country]

# Grafico dell'andamento delle malattie cardiovascolari nel tempo
plt.figure(figsize=(10,5))
plt.plot(df_country["Year"], df_country["Cardiovascular Diseases"], marker="o", label="Cardiovascular Diseases")
plt.xlabel("Anno")
plt.ylabel("Numero di morti")
plt.title(f"Andamento delle malattie cardiovascolari in {country}")
plt.legend()
plt.grid()
plt.show()
```

```python
# Rimuove colonne non numeriche per la correlazione
numeric_df = df.drop(columns=['Country/Territory', 'Code', 'Year'])

# Calcola la matrice di correlazione
correlation_matrix = numeric_df.corr()

# Crea la mappa di calore
plt.figure(figsize=(18, 15))
sns.heatmap(correlation_matrix, cmap='coolwarm', annot=False, fmt=".2f", square=True, linewidths=.5)
plt.title('Mappa di correlazione tra le cause di morte')
plt.tight_layout()
plt.show()
```

```python
# Ottieni l'elenco dei paesi disponibili
countries = df["Country/Territory"].unique()
print("Paesi disponibili:", ", ".join(countries))

# Chiedi all'utente di scegliere un paese
selected_country = "Afghanistan" #input("Scegli un paese tra quelli disponibili: ")

# Filtra i dati per il paese selezionato
if selected_country not in countries:
  print("Paese non valido.")
else:
  df_selected_country = df[df["Country/Territory"] == selected_country]

  # Ottieni gli anni disponibili per il paese selezionato
  years = df_selected_country["Year"].unique()
  print("Anni disponibili:", ", ".join(map(str, years)))

  # Chiedi all'utente di scegliere un anno
  selected_year = 2019 #int(input("Scegli un anno tra quelli disponibili: "))

  if selected_year not in years:
    print("Anno non valido.")
  else:
    # Filtra i dati per l'anno selezionato
    df_selected_year = df_selected_country[df_selected_country["Year"] == selected_year]

    # Trova la causa di morte piÃ¹ probabile
    causes = df_selected_year.drop(columns=["Country/Territory", "Code", "Year"]).iloc[0]
    most_probable_cause = causes.idxmax()
    deaths = causes.max()

    print(f"Nel {selected_year}, in {selected_country}, la causa di morte piÃ¹ probabile Ã¨ stata '{most_probable_cause}' con {deaths} morti.")
```

```python

# Scegli un anno e un paese (es. Italia nel 2019)
anno = 2019
paese = "Italy"

# Filtra i dati per anno e paese
riga = df[(df["Country/Territory"] == paese) & (df["Year"] == anno)]

# Rimuove colonne non legate a malattie
malattie = riga.drop(columns=["Country/Territory", "Code", "Year"])

# Calcola le percentuali
totale_morti = malattie.sum(axis=1).values[0]
percentuali = (malattie / totale_morti * 100).T
percentuali.columns = ["Percentuale"]

# Ordina le cause dalla piÃ¹ alta
percentuali = percentuali.sort_values("Percentuale", ascending=False)

# Grafico
plt.figure(figsize=(12, 8))
percentuali.head(15).plot(kind='barh', legend=False, color='skyblue')  
plt.xlabel("Percentuale di morti (%)")
plt.title(f"Principali cause di morte in {paese} nel {anno}")
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```
